### Обработка пропусков

```python
print(data.isnull().sum()) #подсчет отсутствуюищх значений в каждом столбце
```

```python
print(data['ceiling_height'].value_counts()) #определяем какие значения встречаются в столбце ceiling_height
data[data['ceiling_height'].isna()]
```

```python
#На предыдущем этапе выявлена высота потолка 100 метров, поэтому подробнее рассмотрим данные
data['ceiling_height'].describe()
```

```python
print(data['floors_total'].value_counts()) #определяем какие значения встречаются в столбце floors_total
data[data['floors_total'].isna()]
```

```python
print(data['is_apartment'].value_counts()) #определяем какие значения встречаются в столбце is_apartment
data['is_apartment'] = data['is_apartment'].fillna(False) #заполняем пропущенные значения в стоблце is_apartment на False
print(data['is_apartment'].value_counts()) #проверка количества значений после замены
```

```python
print(data['balcony'].value_counts()) #определяем какие именно и сколько раз встречаются значения столбца balcony
data['balcony'] = data['balcony'].fillna(0) #заполняем пропущенные в стоблце balcony значения на 0
print(data['balcony'].value_counts()) #проверка количества значений после замены
```

```python
from pymystem3 import Mystem #импорт библиотеки Mystem
m = Mystem() #импорт библиотеки Mystem
from collections import Counter #вызов специального контейнера Counter
```

```python
print(data['locality_name'].value_counts().head(15))
```

```python
data.dropna(subset=['locality_name'], inplace=True) #удаление пропущенных значений столбца locality_name
unique_locality = data['locality_name'].unique() #подсчет общей частотности и встречаемости названий населенных пунктов
all_lemmas = []
for locality in unique_locality:
    lemmas = ''.join(m.lemmatize(locality)).strip()
    all_lemmas.append(lemmas)
print(Counter(all_lemmas)) #применение счетчика Counter
```

```python
data['locality_name'] = data['locality_name'].str.replace('ё','е') #замена ё на е в столбце locality_name
```

```python
### удаление из наименования населенного пункта типов населенных пунктов
data['locality_name'] = data['locality_name'].replace(r'городской поселок|деревня|поселок|поселок городского типа|село|городской|садовое товарищество|городского типа|коттеджный|при железнодорожной станции|садоводческое некоммерческое товарищество', '', regex=True)
```

```python
print(data['locality_name'].unique())
```

```python
data['locality_name'] = data['locality_name'].str.lower() # сохраняем столбец locality_name в нижнем регистре
print(data['locality_name'].value_counts().head(15))
```
