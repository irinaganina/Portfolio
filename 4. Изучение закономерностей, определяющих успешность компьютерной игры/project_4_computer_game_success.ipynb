## Шаг 2. Подготовка данных

```python
data.columns = data.columns.str.lower() #замена названий столбцов, приведение к нижнему регистру
data
```

```python
data.dropna(subset=['name'], inplace=True)
```

```python
data.dropna(subset=['genre'], inplace=True)
```

```python
data.info()
```

```python
data['user_score'].value_counts()
```

```python
#замена tbd на пропущенные значения
data['user_score']=data['user_score'].replace('tbd', np.NaN)
```

```python
#замена типов данных
data['year_of_release'] = data['year_of_release'].round().astype('Int64')
data['user_score'] = data['user_score'].astype('float64')
```

```python
#добавление столбца с данными о суммарных продажах во всех регионах
data['total_sales'] = data['na_sales']+ data['eu_sales'] + data['jp_sales']+ data['other_sales']
```

```python
data.info()
```

```python
#изучение значений столбца rating
data['rating'].value_counts()
```

По информации из Википедии (ссылка: https://www.wikiwand.com/ru/Entertainment_Software_Rating_Board):

* «E» («Everyone») — «Для всех» (первоначально "K-A" ("Kids to Adults"))
* «E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»
* «EC» («Early childhood») — «Для детей младшего возраста»

* «T» («Teen») — «Подросткам»

* «M» («Mature») — «Для взрослых»
* «AO» («Adults Only 18+») — «Только для взрослых»

* «RP» («Rating Pending») — «Рейтинг ожидается»

Есть одинаковые по смыслу, но разные по обозначению рейтинги. Их можно объединить в укрупенные группы: E, M и T.

```python
#замена значений
data['rating']=data['rating'].replace('K-A', 'E')
data['rating']=data['rating'].replace('EC', 'E')
data['rating']=data['rating'].replace('E10+', 'E')
data['rating']=data['rating'].replace('AO', 'M')
```

```python
#проверка значений столбца rating после замены
data['rating'].value_counts()
```

## Вывод

Наименования столбцов приведены к нижнему регистру.

Строки, у которых есть пропуски в столбцах *'name'* и *'genre'*, были удалены, поскольку таких значений меньше 10% от всего набора количества строк, поскольку их удаление не приведет к искажению результатов анализа. 

Аббревиатура от английского **to be determined (будет определено)** или **to be decided (будет решено)**. Используется, если какая-то информация еще не определена или решение по вопросу не принято. Поэтому в столбце *'user_score'* значения *'tbd'* были заменены на пропуски. 

Заменен тип данных в столбце *'year_of_release'* с вещественного на целочисленный, поскольку значение года не может быть не целым числом. Тип столбца *'user_score'* заменен со строкового на вещественный, поскольку оценка пользователей принимает значения от 0 до 10.

Платформы с наибольшими суммарными продажами: PS2, X360, PS3, Wii, DS, PS.

```python
data_actual.pivot_table(index='year_of_release', values='total_sales', columns=['platform'], aggfunc='sum').plot(figsize=(15,10))
plt.ylabel("Суммарные gjzпродажи")
plt.xlabel("Год выпуска")
plt.title("Суммарные продажи за весь период")
```

```python
pivot = data_actual.pivot_table(index='year_of_release', values='total_sales', columns=['platform'], aggfunc='sum')
```

```python
pivot
```

```python
pivot.fillna(0, inplace=True)
pivot
```

```python
#поиск средней продолжительности жизни игровой платформы. список из количества лет жизни каждой платформы
notnull=[]
for platform in pivot.columns:
    notnull.append(pivot[platform].ne(0).value_counts().loc[pivot[platform].ne(0).value_counts().index == True].values[0])
```

```python
notnull = pd.DataFrame(notnull, columns=['active_years'])
notnull.describe()
```

```python
sns.boxplot(y=notnull['active_years'])
plt.ylim(9.50,12.75)
```

Таким образом, **средняя продолжительность жизни игровой платформы**, в течение которого она приносит деньги, составляет **11 лет.**

#### Вывод

Наиболее прибыльными платформами в рассматриваемом датасете являются PS2, X360, PS3, Wii, DS, PS.
Практически все игровые платформы существуют 11 лет. Жизненный цикл платформы можно описать следующим образом. Новые платформы достигают пика продаж за 5 лет, а затем продажи идут на спад.

## Шаг 5. Проверка гипотез

Гипотезы:

* средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* средние пользовательские рейтинги жанров Action и Sports разные.

Для проверки обеих гипотез будем использовать t-критерий Стьюдента, т.к. в обоих случаях выборки независимы между собой.
Определим пороговое значение alpha = 0.05

### Гипотеза 1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

```python
actual['platform'].value_counts()
```

**Н0 (нулевая гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC одинаковые.**

**Н1 (альтернативная гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC разные.**

```python
alpha = 0.05#задаем критический уровень статистической значимости
xbox_one_ratings = actual[(actual['platform'] == 'XOne') & (actual['user_score'] > 0)]['user_score']
pc_ratings = actual[(actual['platform'] == 'PC') & (actual['user_score'] > 0)]['user_score']
results = st.ttest_ind(xbox_one_ratings, pc_ratings)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу.")
else:
    print("Не получилось отвергнуть нулевую гипотезу.")
```

#### Вывод
**Не отклоняем нулевую гипотезу.**
Верно следующее утверждение: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

```python
#расчет средних пользовательских рейтингов на платформах
print('Средняя оценка пользователей для платформы Xbox One  - {:.2f}'.format(xbox_one_ratings.mean()))
print('Средняя оценка пользователей для платформы PC - {:.2f}'.format(pc_ratings.mean()))
```

### Гипотеза 2. Средние пользовательские рейтинги жанров Action  и Sports  разные

```python
actual['genre'].value_counts()
```

**Н0 (нулевая гипотеза) - средние пользовательские рейтинги жанров Action и Sports одинаковые.**

**Н1 (альтернативная гипотеза) - средние пользовательские рейтинги жанров Action и Sports разные.**

```python
alpha = 0.05#задаем критический уровень статистической значимости
action_ratings = actual[(actual['genre'] == 'Action') & (actual['user_score'] > 0)]['user_score']
sports_ratings = actual[(actual['genre'] == 'Sports') & (actual['user_score'] > 0)]['user_score']
results = st.ttest_ind(action_ratings, sports_ratings)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу.")
else:
    print("Не получилось отвергнуть нулевую гипотезу.")
```

#### Вывод

**Отклоняем нулевую гипотезу.**
Верно следующее утверждение: средние пользовательские рейтинги платформ Action и Sports отличаются.

На долю самой популярной платформы в Японии приходится 50% продаж от продаж топ-5 платформ в регионе. Лидер по продажам среди платформ в Японии - 3DS. С рейтингом ситуация противоположная американскому и европейскому рынкам. Наиболее продаваемыми являются игры с рейтингом «E» («Everyone») — «Для всех». Затем следуют рейтинг «T» («Teen») — «Подросткам» и рейтинг «M» («Mature») — «Для взрослых».

* При этом важно отметить, что американский рынок больше европейского, поскольку продажи выше. Следует отметить, что японский рынок гораздо меньше по размерам чем европейский и американский.
