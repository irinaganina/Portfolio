### Лемматизация

```python
from pymystem3 import Mystem #импорт библиотеки Mystem
m = Mystem() #импорт библиотеки Mystem
```

```python
print(len(data['purpose'].unique())) #подсчет количества уникальных значений
print(data['purpose'].unique()) #вывод на экран список уникальных значений столбца purpose
```

```python
def create_category_purpose(row):
    lem_purpose = m.lemmatize(row['purpose'])
    if 'жилье' in lem_purpose: #выделяем леммы в столбце purpose
        return 'недвижимость'
    if 'автомобиль' in lem_purpose:
        return 'автомобиль'
    if 'образование' in lem_purpose:
        return 'образование'
    if 'недвижимость' in lem_purpose:
        return 'недвижимость'
    if 'свадьба' in lem_purpose:
        return 'свадьба'
    return 'прочие цели'

data['purpose'] = data.apply(create_category_purpose, axis=1) #заменяем значения столбца purpose лемматизированными значениями
```

```python
def children_grouped(children): #функция, которая оценивает количество детей и распределяет по категориям 
    
    if children == 0: 
        return 'Бездетный'
    if children >= 1 and children <= 2: 
        return 'Есть дети'
    return 'Многодетный'

data['children_grouped'] = data['children'].apply(children_grouped) #добавление нового стоблца в датафрейм
print(data['children_grouped'].value_counts()) #определяем численность новых категорий
```

```python
def income_grouped(income_type): #функция, которая оценивает принадлежность к определенной категории доходов
    
    if income_type == 'предприниматель' or income_type =='компаньон':     
        return 'Высокий доход'
    
    if income_type == 'госслужащий' or income_type =='сотрудник':  
        return 'Средний доход'
    
    return 'Низкий доход'

data['income_grouped'] = data['income_type'].apply(income_grouped) #добавление нового стоблца в датафрейм
print(data['income_grouped'].value_counts()) #определяем численность новых категорий
```
