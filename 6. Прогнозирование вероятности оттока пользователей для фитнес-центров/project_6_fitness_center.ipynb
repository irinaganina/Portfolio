# Прогнозирование вероятности оттока пользователей для фитнес-центров

# Описание проекта

Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных.
Распространённая проблема фитнес-клубов и других сервисов — ***отток клиентов***.

Чтобы бороться с оттоком, отдел по работе с клиентами «Культуриста-датасаентиста» перевёл в электронный вид множество клиентских анкет. ***Задача — провести анализ и подготовить план действий по удержанию клиентов***.

А именно:
* научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
* сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
* проанализировать основные признаки, наиболее сильно влияющие на отток;
* сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:
    1. выделить целевые группы клиентов;
    2. предложить меры по снижению оттока;
    3. определить другие особенности взаимодействия с клиентами.

## Шаг 3. Модель прогнозирования оттока клиентов

### 3.1. Обучающая и валидационная выборка

```python
# разделение данных на признаки (матрица X) и целевую переменную (y)
X = churn.drop('Churn', axis = 1)
y = churn['Churn']
```

```python
# разделение модели на обучающую и валидационную выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### 3.1. Логистическая регрессия

```python
# алгоритм для модели
model = LogisticRegression(solver='liblinear') 
```

```python
# обучение модели
model.fit(X_train, y_train)
```

```python
#прогнозы
predictions = model.predict(X_test) 
probabilities = model.predict_proba(X_test)[:,1]
```

```python
#метрики
print('Accuracy {:.2f}'.format(accuracy_score(y_test, predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, predictions)))
```

#### Вывод

Логистическая модель прогнозирует правильно 93% ответов.
Метрики точность модели составляет 86%.
Полнота модели составляет составляет 83%. 
Метрики указывают на хорошее качество модели.

### 3.2. Cлучайный лес

```python
# зададим алгоритм для новой модели на основе алгоритма случайного леса
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0) 
```

```python
# обучим модель случайного леса
rf_model.fit(X_train, y_train)
```

```python
# воспользуемся уже обученной моделью, чтобы сделать прогнозы
rf_predictions = rf_model.predict(X_test) # Ваш код здесь
rf_probabilities = rf_model.predict_proba(X_test)[:,1]# Ваш код здесь
```

```python
#метрики
print('Accuracy {:.2f}'.format(accuracy_score(y_test, rf_predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, rf_predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, rf_predictions)))
```

#### Вывод

Логистическая модель прогнозирует правильно 92% ответов.
Метрики точность модели составляет 84%.
Полнота модели составляет составляет 82%. 
Метрики указывают на хорошее качество модели, однако по метрикам логистическая модель лучше.

## Шаг 4. Кластеризация клиентов

### 4.1. Стандартизация данных и дендрограмма

```python
# стандартизируем данные
sc = StandardScaler()
x_sc = sc.fit_transform(X)
```

```python
# матрица расстояний
linked = linkage(x_sc, method = 'ward') 
```

```python
#построение дендрограммы
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.show()
```

#### Вывод

Предложенное оптимальное число кластеров 5 — пять разных цветов на графике.

### 4.2. Алгоритм K-Means. Средние значения для признаков. Доля оттока.

```python
# задаём модель k_means с числом кластеров 5
km = KMeans(n_clusters = 5, random_state = 0)
```

```python
# прогнозируем кластеры для наблюдений
labels = km.fit_predict(x_sc)
```

```python
# сохраняем метки кластера в поле нашего датасета
churn['cluster_km'] = labels
```
